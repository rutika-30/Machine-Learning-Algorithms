##### First we will create a class and define '__init__' method to recall/initilize parameters

```python
class animal():
    
    def __init__(self):
        
        print("Hello ")
```

##### Create an instance of class, to use them;

```python
x = animal()
```

##### Hoorayy! we ran our first self created class & it's instance 

##### Moving forwardd... 

##### WHat if we want to add parameteres in the class??

```python
class take_param():
    
    def __init__(self):
        
        print("Hello! ")
        
    def shout(self,name):
        
        print("You cannot shout " + name )
```

```python
y = take_param()
```

##### What if i want to print sound of that animal

```python
y.shout('Crow')
```

##### Now as we have done with class, let's create an extended class, which will inherit features from the existing one!

```python
class extended_class(take_param):
    
    def __init__(self):
        
        print("Welcome to Python land!!")
        
```

```python
s = extended_class()
```

```python
s.shout("Crow")
```

##### Ohho!!! we can see that even if extended class is inherited one, yet we can't print hello

##### and for the same we need to bring in "SUPER" keyword

```python
class extended_class(take_param):
    
    def __init__(self):
        
        super().__init__()
        
        print("Welcome to Python land!!")
        
```

##### SUPER keyowrd enables the 'init' function of parent class to get executed here

##### So, let's check it out!


```python
new = extended_class()
```

```python
new.shout('Minty')
```
